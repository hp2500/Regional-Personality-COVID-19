```python
import pandas as pd
from itertools import chain
```

```python
dict_pickle = pd.read_pickle('/Users/hp2500/Google Drive/STUDY/Columbia/Research/Corona/Data/features.p')

df_temp = pd.DataFrame()
for i in dict_pickle.items():
    for j in i[1].items():
        row_temp = pd.Series(chain([i[0]], [j[0]], j[1]))
        df_temp = df_temp.append(row_temp, ignore_index=True)
```

```python
col_names = ['uid', 'date', 'loc_var_value', 'dis_value', 'speed_mean', 'speed_var', 
 'convex_hull_value', 'transition_time_value', 'dis_ent_value', 'raw_ent_value', 
 'dis_var', 'dis_mean', 'max_dis_from_home', 'max_dis', 'num_gps_records', 
 'num_cluster_value', 'time_at_each_loc_value', 'location_changes_value', 
 'ent_value', 'norm_ent_value', 'rad_gyration_value', 'per_1', 'per_2', 'per_3', 
 'per_cluster','per_at_home', 'q5', 'q10', 'q25', 'q50', 'q75', 'q90', 'q95', 
 'iqr', 'skew']
 
 
```

```python
df_temp.columns = col_names
```
